```python
# ===== ONE-HOT ENCODING COMPLETE GUIDE =====

# Example 1: Basic One-Hot Encoding with Colors
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np

print("=" * 50)
print("EXAMPLE 1: Basic Color Encoding")
print("=" * 50)

colors_data = [['Red'], ['Blue'], ['Green'], ['Red'], ['Blue']]
encoder = OneHotEncoder(sparse_output=False)
encoded_colors = encoder.fit_transform(colors_data)

print("Original colors:", [c[0] for c in colors_data])
print("Category names:", encoder.categories_[0])
print("One-Hot Encoded:\n", encoded_colors)
df = pd.DataFrame(encoded_colors, columns=encoder.categories_[0])
print("As DataFrame:\n", df)
```

```python
# ===== ONE-HOT ENCODING WITH TEXT =====
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
import numpy as np

# Your text data
text = ["prithwi is good", "i am bad", "prithwi is bad"]

# Step 1: Extract all unique words
all_words = []
for sentence in text:
    words = sentence.lower().split()
    all_words.extend(words)

unique_words = list(set(all_words))
print("Unique words:", sorted(unique_words))

# Step 2: Convert text to word vectors (list of words for each sentence)
text_vectors = []
for sentence in text:
    words = sentence.lower().split()
    text_vectors.append(words)

# Step 3: Create and fit OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)

# Prepare data: each word as a separate row
encoded_data = []
for words in text_vectors:
    for word in words:
        encoded_data.append([word])

# Fit and transform
encoded = encoder.fit_transform(encoded_data)

print("\n" + "="*50)
print("ONE-HOT ENCODING OUTPUT")
print("="*50)
print("\nOriginal text:")
for i, t in enumerate(text):
    print(f"  {i+1}. {t}")

print("\nWord categories:", encoder.categories_[0])

print("\nEncoded shape:", encoded.shape)
print("\nEncoded matrix (each row = one word):")
print(encoded)

# Create DataFrame
df_encoded = pd.DataFrame(encoded, 
                          columns=encoder.categories_[0])
print("\nAs DataFrame:")
print(df_encoded)

print("\nExplanation:")
print("- Each row = one word")
print("- Each column = one unique word category")
print("- 1 = this is the word")
print("- 0 = this is not the word")
```
